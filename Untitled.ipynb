```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(12) + 1
y = np.array([60 + 55.898, 44.130, 31.854, 20.913, 18.843, 16.240, 16.054, 14.858, 12.989, 11.746, 12.048, 11.428])
```

```python
new_y = y[0]/y
#new_y
```

```python
plt.figure(figsize=(15,7))
plt.plot(x, new_y, '-r')
plt.plot(x, x, '-b')
plt.grid()
```

```python
new_y1 = y[0]/y
for i in range(12):
    new_y1[i] /= i + 1
#new_y1
```

```python
plt.figure(figsize=(15,7))
plt.plot(x, new_y1, '-r')
#plt.plot(x, x, '-b')
plt.grid()
```
